```python
import praw
import matplotlib.pyplot as plt
import re
from collections import defaultdict
```

```python
secretsfile = open('secrets.txt', 'r')
secrets = {}
for l in secretsfile:
    kv = l.split('=')
    secrets[kv[0]] = kv[1].strip()
```

```python
# Get ready to call reddit

clientid = secrets['client_id']
clientsecret = secrets['client_secret']
reddit = praw.Reddit(client_id=clientid, client_secret=clientsecret, user_agent="user_agent")
```

```python
# This is the subreddit we want

subreddit = reddit.subreddit("GWASapphic")
```

```python
#Get some posts

posts = []
for post in subreddit.new(limit=500):
    posts.append(post)
```

```python
# We care about the titles of the posts

titles = []
for p in posts:
    titles.append(p.title)
```

```python
# This is how we're going to get the tags from the titles

def parse(title):
    ret = []
    start = 0
    while (start < len(title)):
        if title[start] == '[': # we're in a tag
            end = start + 1
            while end < len(title) and (title[end] != ']' and title[end] != '['):
                end = end + 1 # tag continues
            # tag has ended, add it to the list (in lowercase)
            ret.append(title[start+1:end].lower())
            start = end + 1
            end = end + 1
        else:
            start = start + 1 # outside a tag, keep looking
    return ret
                
```

```python
# Shape the data how we want it

# More structured data for each post
postdata = []

# Occurrence counts for gender and post type tags
allgenders = defaultdict(int)
alltypes = defaultdict(int)

# Occurrence counts for all other tags
alltags = defaultdict(int)

# Populate these counts
for p in posttags:
    entry = {"genders": [], "types": [], "tags": []}
    for t in p:
        if re.search(r'[a-z]+4[a-z]+', t): # Look for gender tags
            entry["genders"].append(re.sub(r'\W', '', t).upper()) # Strip special characters and uppercase the tag
            allgenders[t.upper()] = allgenders[t.upper()] + 1
        elif "script" in t or "request" in t:
            entry["types"].append(re.sub(r'\W', ' ', t)) # Replace special characters with space
            alltypes[t] = alltypes[t] + 1
        else:
            entry["tags"].append(t)
            alltags[t] = alltags[t] + 1
    # If we're missing a gender or type, count it as 'unknown'
    if len(entry["types"]) == 0:
        alltypes["unknown"] = alltypes["unknown"] + 1
    if len(entry["genders"]) == 0:
        allgenders["unknown"] = allgenders["unknown"] + 1
    entry["genders"] = sorted(entry["genders"])
    postdata.append(entry)

print(postdata[:3])
```

```python
# Sort the data, larges values first

sortedtags = {k: v for k, v in sorted(alltags.items(), key=lambda item: -item[1])}
sortedtypes = {k: v for k, v in sorted(alltypes.items(), key=lambda item: -item[1])}
sortedgenders = {k: v for k, v in sorted(allgenders.items(), key=lambda item: -item[1])}

print("found %d post types" % len(sortedtypes))
print("found %d genders" % len(sortedgenders))
print("found %d other tags" % len(sortedtags))
```

```python
# This is how we're going to draw the graphs
# Limit the number of entries for each graph so they're as readable as possible

def plot(inputdata, limit, title):
    plt.figure(figsize=(20,12))
    ax = plt.subplot()
    xs = []
    ys = []
    xs.extend(inputdata.keys())
    ys.extend(inputdata.values())
    ax.invert_yaxis()
    ax.set_title(title)
    ax.barh(xs[:limit], ys[:limit])
```
